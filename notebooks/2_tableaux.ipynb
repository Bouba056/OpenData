Tableau 1 : Evolution des logements selon le type de l'habitat

```python
df.head()
```

```python
histo_logement = df[["LIBGEO", "AN", "LOG", "MAISON", "APPART"]].copy()
```

```python
# Pivot longer (√©quivalent de pivot_longer)
histo_logement = histo_logement.melt(
    id_vars=["LIBGEO", "AN", "LOG"],
    value_vars=["APPART", "MAISON"],
    var_name="TYPE_HABITAT",
    value_name="NOMBRE"
)

# Remplacement des valeurs dans TYPE_HABITAT
histo_logement["TYPE_HABITAT"] = histo_logement["TYPE_HABITAT"].replace({
    "APPART": "Appartements",
    "MAISON": "Maisons"
})
```

```python
histo_logement["LOG"] = histo_logement["LOG"].round(0).astype(int)
histo_logement["NOMBRE"] = histo_logement["NOMBRE"].round(0).astype(int)
```

```python
histo_logement.head()
```

```python
histo_logement.to_csv("../SORTIE/TAB_TYPEHAB.csv", index=False)
```

Tableau 2: evolution selon la categorie du logement

```python
histo_logement_cate = df[["LIBGEO", "AN", "RP", "RSECOCC", "LOGVAC"]].copy()

# Pivot longer
histo_logement_cate = histo_logement_cate.melt(
    id_vars=["LIBGEO", "AN"],
    value_vars=["RP", "RSECOCC", "LOGVAC"],
    var_name="TYPE_LOG",
    value_name="NOMBRE"
)

# Remplacer les noms de cat√©gories
histo_logement_cate["TYPE_LOG"] = histo_logement_cate["TYPE_LOG"].replace({
    "RP": "R√©sidences principales",
    "RSECOCC": "R√©sidences secondaires et occasionnelles",
    "LOGVAC": "Logements vacants"
})
```

```python
histo_logement_cate["NOMBRE"] = histo_logement_cate["NOMBRE"].round(0).astype(int)
```

```python
histo_logement_cate.head()
```

```python
histo_logement_cate.to_csv("../SORTIE/TAB_CATEHAB.csv", index=False)
```

Tableau 3 : r√©sidences principales selon le nombre de pieces 

```python
# D√©finition des colonnes √† pivoter
cols_typo_log = ["RP_1P", "RP_2P", "RP_3P", "RP_4P", "RP_5PP"]

# S√©lection + pivot_longer
rp_typo = df[["LIBGEO", "AN"] + cols_typo_log].melt(
    id_vars=["LIBGEO", "AN"],
    value_vars=cols_typo_log,
    var_name="TYPO",
    value_name="NOMBRE"
)

# Remplacement des libell√©s
rp_typo["TYPO"] = rp_typo["TYPO"].replace({
    "RP_1P": "T1",
    "RP_2P": "T2",
    "RP_3P": "T3",
    "RP_4P": "T4",
    "RP_5PP": "T5et+"
})
```

```python

rp_typo["NOMBRE"] = rp_typo["NOMBRE"].round(0).astype(int)
```

```python
rp_typo.head()
```

```python
rp_typo.to_csv("../SORTIE/RP_TYPO.csv", index=False)
```

Tableau 4 : r√©sidences principales selon leut p√©riodes de construction

```python

rp_cons1 = df[df["AN"] == df["AN"].max()][
    ["LIBGEO", "AN", "RP_ACH19", "RP_ACH45", "RP_ACH70", "RP_ACH90", "RP_ACH05", "RP_ACH18"]
].copy()

cons_date = [c for c in rp_cons1.columns if c not in ["AN", "LIBGEO"]]

rp_cons = rp_cons1.melt(
    id_vars=["LIBGEO", "AN"],
    value_vars=cons_date,
    var_name="ANNEE_CONS",
    value_name="NOMBRE"
)

rp_cons["ANNEE_CONS"] = rp_cons["ANNEE_CONS"].replace({
    "RP_ACH19": "Construites avant 1919",
    "RP_ACH45": "Construites de 1919 √† 1945",
    "RP_ACH70": "Construites de 1946 √† 1970",
    "RP_ACH90": "Construites de 1971 √† 1990",
    "RP_ACH05": "Construites de 1991 √† 2005",
    "RP_ACH18": "Construites de 2006 √† 2018"
})
```

```python
rp_cons.head()
```

```python
pd.crosstab(rp_cons["ANNEE_CONS"], rp_cons["NOMBRE"])
```

```python
rp_cons["ANNEE_CONS"].value_counts()
```

```python
rp_cons.to_csv("../SORTIE/RP_CONS.csv", index=False)
```

Tableau 4 : rp selon le statut d'occupation

```python
cols_so_rp = ["RP_LOCHLMV", "RP_LOCPRIV", "RP_PROP", "RP_GRAT"]

# S√©lection + pivot_longer (melt)
rp_so = df[["AN", "LIBGEO"] + cols_so_rp].melt(
    id_vars=["AN", "LIBGEO"],
    value_vars=cols_so_rp,
    var_name="STATUT",
    value_name="NOMBRE"
)

# Filtrer sur la derni√®re ann√©e
rp_so = rp_so[rp_so["AN"] == rp_so["AN"].max()]

# Remplacer les labels du statut
rp_so["STATUT"] = rp_so["STATUT"].replace({
    "RP_LOCHLMV": "Locataire HLM (logement lou√© vide)",
    "RP_LOCPRIV": "Locataire du parc priv√© (locataire hors logement HLM lou√© vide)",
    "RP_PROP": "Propri√©taire",
    "RP_GRAT": "Log√© gratuitement"
})
```

```python
rp_so["NOMBRE"] = rp_so["NOMBRE"].round(0).astype(int)
```

```python
rp_so.head()
```

```python
rp_so.to_csv("../SORTIE/RP_SO.csv", index=False)
```
